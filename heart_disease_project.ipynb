## Data Description
The following is a brief description on what information the following columns provide:
**Response Variable:**
* `HeartDiseaseorAttack`
*   Respondents that have ever reported having coronary heart disease (CHD) or myocardial infarction (MI). True or false


**Independent Variables:**

**High Blood Pressure**
* `HighBP`
*   Adults who have been told they have high blood pressure by a doctor, nurse, or other health professional. True or false

**High Cholesterol**
* `HighChol`
*   Have you EVER been told by a doctor, nurse or other health professional that your blood cholesterol is high?
* `CholCheck`
*   Cholesterol check within past five years. True or false

**BMI**
* `BMI`
*   Body Mass Index (BMI)

**Smoking**
* `Smoker`
*   Have you smoked at least 100 cigarettes in your entire life? [Note: 5 packs = 100 cigarettes]. True or false

**Other Chronic Health Conditions**
* `Stroke`
*   Ever told you had a stroke. True or false
* `Diabetes`
*   Ever told you have diabetes. "False" = 0, 0 is for no diabetes or only during pregnancy, 1 is for pre-diabetes or borderline diabetes, 2 is for yes diabetes

**Physical Activity**
* `PhysActivity`
*   Adults who reported doing physical activity or exercise during the past 30 days other than their regular job. True or false

**Diet**
* `Fruits`
*   Consume Fruit 1 or more times per day. True or false
* `Veggies`
*   Consume Vegetables 1 or more times per day. True or false

**Alcohol Consumption**
* `HvyAlcoholConsump`
*   Heavy drinkers . True or false

**Health Care**
* `AnyHealthcare`
*   Do you have any kind of health care coverage, including health insurance, prepaid plans such as HMOs, or government plans such as Medicare, or Indian Health Service? True or false
* `NoDocbcCost`
*   Was there a time in the past 12 months when you needed to see a doctor but could not because of cost? True or false

**Health General and Mental Health**
* `GenHlth`
*   Would you say that in general your health is: a scale of 1 to 5.(1 is Excellent -> 5 is Poor)
* `MentHlth`
*   Now thinking about your mental health, which includes stress, depression, and problems with emotions, for how many days during the past 30 days was your mental health not good?
* `PhysHlth`
*   Now thinking about your physical health, which includes physical illness and injury, for how many days during the past 30 days was your physical health not good?
* `DiffWalk`
*   Do you have serious difficulty walking or climbing stairs? True or false

**Demographics**
* `Sex`
*   Indicate sex of respondent. 1-male, 0-female
* `Age`
*   Fourteen-level age category;

    1 -- 18 to 24 years

    2 -- 25 to 29 years

    3 -- 30 to 34 years

    4 -- 35 to 39 years

    5 -- 40 to 44 years

    6 -- 45 to 49 years

    7 -- 50 to 54 years

    8 -- 55 to 59 years

    9 -- 60 to 64 years

    10 -- 65 to 69 years

    11 -- 70 to 74 years

    12 -- 75 to 79 years

    13 -- 80 years or older 
* `Education`   
*   What is the highest grade or year of school you completed? 1 being never attended school or kindergarten only up to 6 being college 4 years or more
* `Income`
*   Is your annual household income from all sources: 0 is no and Variable is already ordinal with 1 being less than $10,000 all the way up to 8 being $75,000 or more

```python
# Creating a function to draw count plots
def draw_counts(data, column, target:None):
    plt.figure(figsize=(16,8))
    plt.suptitle("Plot value count", fontsize=20, x=0.5, y=1)
    sns.countplot(data=heart_data, x=column, hue=target)
    plt.title(f"{column}")
    plt.show()
 
# creating a function to draw pie charts 
def get_pie(data, column):
    plt.figure(figsize=(16,8))
    plt.suptitle("Plot value count", fontsize=20, x=0.5, y=1)
    plt.pie(x=heart_data[column].value_counts(), labels=heart_data[column].unique(), autopct='%.0f%%')
    plt.title(f"{column}")
    plt.show()

# creating a function that draws distribution plots   
def get_distplot(data, column):
    plt.figure(figsize=(16,8))
    plt.suptitle("Plot distribution", fontsize=20, x=0.5, y=1)
    sns.distplot(x=heart_data[column], kde=True, color='red', bins=100)
    plt.title(f"{column}")
    plt.show()
    
# creating a function to draw bar plot to show relationships of continuous and categorical variables
def get_barplot(data, column, target):
    plt.figure(figsize=(16,8))
    plt.suptitle("Plot value count", fontsize=20, x=0.5, y=1)
    sns.barplot(data=heart_data, x=target, y=column)
    plt.title(f"{column}")
    plt.show()
    
# creating a function that draws scatter plots to draw relationships between continuous variables
def get_scatter(data, x_val, y_val):
    plt.suptitle("Scatter plot", fontsize=20, x=0.5, y=1)
    sns.scatterplot(data=heart_data, x=x_val, y=y_val)
    plt.title(f"{x_val} Vs {y_val}")
    plt.show()
```

```python
# plotting a correlation matrix
matrix = np.zeros_like(heart_data.corr())
matrix[np.triu_indices_from(matrix)] = True

with sns.axes_style("white"):
    fig, ax = plt.subplots(figsize = (15, 15))
    ax = sns.heatmap(heart_data.corr(), 
                     annot = False,
                     annot_kws={"size": 8},
                     ax = ax, 
                     cbar_kws = {"shrink": .5}, 
                     mask = matrix, 
                     vmax = 1,
                     vmin = -1,
                     square = True)
```

```python
#checking  high correlatiom, above 0.75
#check for multicollinearity
high_corr = np.zeros_like(abs(heart_data.corr()) > 0.75)
high_corr[np.triu_indices_from(matrix)] = True

with sns.axes_style("white"):
    fig, ax = plt.subplots(figsize = (15, 15))
    ax = sns.heatmap(abs(heart_data.corr()) > 0.75, 
                     annot = True, 
                     ax = ax, 
                     cbar_kws = {"shrink": .5}, 
                     mask = high_corr, 
                     vmax = 1,
                     vmin = -1,
                     square = True)
```

```python
# getting correlation between the various features
df_corr = heart_data.corr().abs().stack().reset_index().sort_values(0, ascending = False)

# listing the pairs
df_corr['pairs'] = list(zip(df_corr['level_0'], df_corr['level_1']))
df_corr.set_index(['pairs'], inplace = True)

#dropping the level_i columns
df_corr.drop(columns = ['level_0', 'level_1'], inplace = True)

#add correlation column
df_corr.columns = ['cc']

#dropping duplicates
df_corr.drop_duplicates(inplace = True)

#correlations over .75, but less than 1. = multicollinarity
df_corr[(df_corr['cc'] > 0.75) & (df_corr['cc'] < 1)]
df_corr.head(20)
```

#### Observations:
A high correlation exists between `BMI` and `Weight`.
